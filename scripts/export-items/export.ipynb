```python
import requests
import json
import re

items = requests.get('http://adapt2.sis.pitt.edu/pcex-authoring/api/hub?key=').json()
print(f"loading items...")

exported_items = []
for item_summary in items:
    if item_summary['author']['email'] in ['ahromic1@etf.unsa.ba', 'mkamberovi1@etf.unsa.ba']:
       print(f"exporting item {item_summary['id']} - {item_summary['name']}")
       
       if len(item_summary['items']) < 1:
            print("  no activity goals found, skipping")
            continue
       
       try:
            item = requests.get(f"http://adapt2.sis.pitt.edu/pcex-authoring/api/hub/{item_summary['id']}?_t=1768393349443").json()[0]
            for goal in item['activityGoals']:
                print(f"  with goal {goal['id']} - {goal['name']}")
                exported_items.append({
                        'id': goal['id'],
                        'activityName': goal['goalDescription'],
                        'name': re.sub(r"__[0-9a-fA-F]+-[0-9a-fA-F]+-\d+", "", goal['name']),
                        'language': goal['language'],
                        'goalDescription': goal['goalDescription'],
                        'userInput': goal['userInput'],
                        'correctOutput': goal['correctOutput'],
                        'fullyWorkedOut': goal['fullyWorkedOut'],
                        'sourceCode': '\n'.join([line['content'].strip() for line in sorted(goal['lineList'], key=lambda x: x['number'])]),
                })
       except Exception as e:
            print(f"  error exporting item: {e}")
            continue

json.dump(exported_items, open('exported_items.json', 'w'), indent=2)
```
