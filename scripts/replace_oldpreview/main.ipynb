```python
import csv

aggregate = {id: url for [id, url] in csv.reader(open('aggregate.csv', 'r'))}
um2 = {id: url for [id, url] in csv.reader(open('um2.csv', 'r'))}
acos = open('acos.txt', 'r').readlines()
```

```python
acos_wid2link = {each.strip().split('__')[1]: each for each in acos}
```

```python
for id, url in aggregate.items():
    # https://pawscomp2.sis.pitt.edu/pcex-authoring/assets/preview/index.html?load=https://pawscomp2.sis.pitt.edu/pcex-authoring/api/hub/664e4f1091363872f0ba38ae?_t=1724341504318
    wid = url.split('/pcex-authoring/api/hub/')[1].split('?_t=')[0]
    if wid not in acos_wid2link:
        print(f'# NOT FOUND: ID={id}, WID={wid}, URL={url}')
        continue
    aggregate[id] = acos_wid2link[wid].strip()
    print(f"UPDATE ent_content SET url = '{aggregate[id]}' WHERE content_id = {id};")
```

```python
for id, url in um2.items():
    wid = url.split('/pcex-authoring/api/hub/')[1].split('?_t=')[0]
    if wid not in acos_wid2link:
        print(f'# NOT FOUND: ID={id}, WID={wid}, URL={url}')
        continue
    um2[id] = acos_wid2link[wid].strip()
    print(f"UPDATE ent_activity SET URI = '{um2[id]}' WHERE ActivityID = {id};")
```
