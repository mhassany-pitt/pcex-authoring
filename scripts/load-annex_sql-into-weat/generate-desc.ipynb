```python
from openai import OpenAI

api_key = open("openai_api_key.txt").read().strip()
client = OpenAI(api_key=api_key)

def generate(sql_query):
    prompt = f"Generate a problem statement (in plain text) such that the correct solution would be the following SQL query:\n\n{sql_query}"

    response = client.responses.create(
        model="gpt-5.4",
        input=[{"role": "user","content": [{"type": "input_text","text": prompt}]},],
        text={"format": {"type": "text"},"verbosity": "low"},
        reasoning={"effort": "medium","summary": "auto"},
        store=False
    )
    
    return response.model, response.output_text
```

```python
import json

items = json.load(open("weat_sql.json"))
for index, item in enumerate(items):
    if '__desc_gpt_generated' not in item:
        print(f"[{index}/{len(items)}] Generating description for {item['name']}...")
        model, desc = generate(item['code'])
        item['description'] = desc
        item['__desc_gpt_generated'] = True
        item['tags'].append("desc-generated-by:" + model)
        json.dump(items, open("weat_sql.json", "w"), indent=2)
```
