```python
# this script fixes the content_names for pcex_examples in ent_content table
# it overwrites the content_name with activity_name for pcex_examples

import csv
import json
import requests

file = open("ent_contents.csv", mode="r")
ent_contents = []
for row in csv.DictReader(file):
    ent_contents.append(row)
file.close()

for ent_content in ent_contents:
    url = ent_content["url"]
    if "index=0" not in url:
        continue

    url = (
        ent_content["url"]
        .replace(
            "http://pawscomp2.sis.pitt.edu/pcex-authoring/preview/index.html?load=", ""
        )
        .split("?")[0]
    )

    if ent_content["creator_id"] != "moh70@pitt.edu":
        continue

    activity = requests.get(url).json()
    if (
        type(activity) is list
        and activity[0]["activityName"] != ent_content["content_name"]
    ):
        print(
            f"UPDATE ent_content SET content_name = '{activity[0]['activityName']}' WHERE content_id = {ent_content['content_id']};"
        )
```
