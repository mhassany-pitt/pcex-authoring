```python
import os
import json
import requests

courses = json.loads(open("list.json").read())

pcex_links = set()
for course_name in courses:
    course_items = courses[course_name].values()
    for pcex_link in course_items:
        if "&ch=" in pcex_link:
            pcex_link = pcex_link.split("&ch=")[0]
        pcex_filename = pcex_link.split("?")[1][5:].replace("&set=", "_")

        if os.path.exists(f"pcex_v1/{pcex_filename}.json"):
            print("skipped!", pcex_filename)
            continue

        pcex_link = f"http://pawscomp2.sis.pitt.edu/pcex/pcex_v1/data/{pcex_filename}.json?v=202405301450"
        pcex_content = requests.get(pcex_link).json()

        json.dump(pcex_content, open(f"pcex_v1/{pcex_filename}.json", "w"), indent=2)
        print("done!", pcex_link)
```

```python
hub = json.loads(open("hub.json").read())

for pcex_item in hub:
    pcex_link = f"http://adapt2.sis.pitt.edu/pcex-authoring/api/hub/{pcex_item['id']}?_t=1760640405677"

    if os.path.exists(f"hub/{pcex_item['id']}.json"):
        print("skipped!", pcex_item["id"])
        continue

    pcex_content = requests.get(pcex_link).json()
    json.dump(pcex_content, open(f"hub/{pcex_item['id']}.json", "w"), indent=2)
    print("done!", pcex_link)
```

```python
pcex_v1 = dict()
for filename in os.listdir("pcex_v1"):
    if filename.endswith(".json"):
        pcex_content = json.loads(open(f"pcex_v1/{filename}").read())
        if pcex_content["activityName"] in pcex_v1:
            print("duplicate!", pcex_content["activityName"])
        pcex_v1[pcex_content["activityName"]] = pcex_content
```

```python
hub = dict()
for filename in os.listdir("hub"):
    if filename.endswith(".json"):
        pcex_content = json.loads(open(f"hub/{filename}").read())[0]
        if pcex_content["activityName"] in hub:
            print("duplicate!", pcex_content["activityName"])
        pcex_content["hub_id"] = filename[:-5]
        hub[pcex_content["activityName"]] = pcex_content
```

```python
matched_activity_names = list(pcex_v1.keys() & hub.keys())
# unmatched_activity_names = list(pcex_v1.keys() - hub.keys())
# print("unmatched:", unmatched_activity_names)
matched_activity_names += [
    "py_pythagorean_theorem__664e23fd91363872f0ba32ea",
    "py_vending_machine__664e24f091363872f0ba3330",
    "artihmetic.vending_machine__664e3efd91363872f0ba35f8",
    "py_time_conversion__664e241291363872f0ba32f4",
    "arithmetic.time_conversion__664e505691363872f0ba38f8",
    "files.j_work_hours__68bafbe6fce5207cc0b4d6ac",
]

print("total matched activities:", len(matched_activity_names))
```

```python
source_names = [
    f"{goal["language"]}--{goal["name"]}"
    for activity_name in matched_activity_names
    for goal in hub[activity_name]["activityGoals"]
]

print("total sources:", len(source_names))
print("total sources (unique):", len(set(source_names)))
print("--------------------------------")
print("\n".join(sorted(list(source_names))))
```

```python
print("hub IDs of matched activities:")
print(
    "  ",
    "\n   ".join(
        sorted(
            [
                f"{hub[an]['language']}-{hub[an]['hub_id']}"
                for an in matched_activity_names
            ]
        )
    ),
)
```
